# Visualization

This notebook contains examples of decision tree visualization using following libraries:
- __graphviz__
- __dtreeviz__

```python
import pickle
from sklearn import tree

with open('reports/model.pkl', 'rb') as f:
    clf = pickle.load(f)
    
with open('reports/train_data.pkl', 'rb') as f:
    X_train = pickle.load(f)
    y_train = pickle.load(f)

with open('reports/feature_data.pkl', 'rb') as f:
    fn = pickle.load(f)
    cn = pickle.load(f)
    dl = pickle.load(f)
```

## Visualization of an example tree's first three levels using graphviz

```python
import graphviz

dot_file = tree.export_graphviz(clf.estimators_[0], max_depth=2, feature_names=fn, class_names=dl, filled=True, rounded=True)
graph = graphviz.Source(dot_file)
graph
```

## Visualization of an example tree using graphviz

```python
dot_file = tree.export_graphviz(clf.estimators_[0], feature_names=fn, class_names=dl, filled=True, rounded=True)
graph = graphviz.Source(dot_file)
graph
```

## Visualization of an example tree's first three levels using dtreeviz

```python
import dtreeviz

viz = dtreeviz.trees.model(clf.estimators_[0], X_train, y_train.values.ravel(), target_name='Heart disease presence', feature_names=fn, class_names=dl)
graph = viz.view(depth_range_to_display=(0, 2), scale=1.3)
graph
```

## Visualization of an example tree using dtreeviz

```python
viz = dtreeviz.trees.model(clf.estimators_[0], X_train, y_train.values.ravel(), target_name='Heart disease presence', feature_names=fn, class_names=dl)
viz.view(scale=1.3)
```
